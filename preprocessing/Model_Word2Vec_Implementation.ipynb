# **Word2Vec Skipgram Model**

## Build Model

```python
import pickle
```

```python
skipgram_model_path = "skipgram_wiki_word2vec_gensim3.pickle"
```

```python
skipgram_model_path = "skipgram_wiki_word2vec_gensim3.pickle"
sentences = gensim.models.word2vec.LineSentence(file_idwiki)
skipgram_model = gensim.models.word2vec.Word2Vec(sentences, vector_size=300, window=5, min_count=1, workers=4, sg=1)
```

```python
with open(skipgram_model_path, 'wb') as f:
    pickle.dump(skipgram_model, f)
```

## Testing

```python
with open(skipgram_model_path, 'rb') as f:
    skipgram_model = pickle.load(f)
```

```python
print(skipgram_model.wv['indonesia'])
```

```python
similarity = skipgram_model.wv.most_similar("komputer", topn=10)
print(similarity)
```
